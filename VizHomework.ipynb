```python
import pandas as pd
import matplotlib.pyplot as plot


data = pd.read_csv('Top_Highest_Openings.csv')


data['Year'] = pd.to_datetime(data['Date'], format='%d-%m-%Y').dt.year #getting the year from date column



plot.figure(figsize=(10, 6)) #scatter plot - total gross vs. year
plot.scatter(data['Year'], data['Total Gross'], color='blue', alpha=0.5)
plot.title('Movie Total Gross Earnings by Year')
plot.xlabel('Year')
plot.ylabel('Total Gross Earnings ($)')
plot.grid(True)
plot.show()


top_10_movies = data.nlargest(10, 'Total Gross')#bar chart - top 10 movies by total gross
plot.figure(figsize=(12, 8))
plot.bar(top_10_movies['Release'], top_10_movies['Total Gross'], color='green')
plot.title('Top 10 Movies by Total Gross Earnings')
plot.xlabel('Movie')
plot.xticks(rotation=45)
plot.ylabel('Total Gross Earnings ($)')
plot.show()


average_gross_by_year = data.groupby('Year')['Total Gross'].mean() #line chart - average total gross by ear
plot.figure(figsize=(12, 8))
average_gross_by_year.plot(kind='line', marker='o', linestyle='-', color='red')
plot.title('Average Movie Total Gross Earnings by Year')
plot.xlabel('Year')
plot.ylabel('Average Total Gross Earnings ($)')
plot.grid(True)
plot.show()
```
